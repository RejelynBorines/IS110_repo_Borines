#### Exercise 2

Write a program to prompt the user for hours and rate per hour to compute gross pay.
```
Enter Hours: 35
Enter Rate: 2.75
Pay: 96.25
```

[PEW] E2.3

````python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = hours * rate
print("Pay:", pay)
````

#### Exercise 3

Assume that we execute the following assignment statements:
```python
width = 17
height = 12.0
```
For each of the following expressions, write the value of the expression and the type (of the value of the expression).
```python
width//2
width/2.0
height/3
1 + 2 * 5
```

[PEW] E2.4

````python
width = 17
height = 12.0


print("width//2 =", width//2, type(width//2))
print("width/2.0 =", width/2.0, type(width/2.0))
print("height/3 =", height/3, type(height/3))
print("1 + 2 * 5 =", 1 + 2 * 5, type(1 + 2 * 5))
````

#### Exercise 4

Write a program that uses input to prompt a user for a number and then outputs the square of the number
```
Enter a number: 3
The square of 3 is 9
```

````python
number = int(input("Enter a number: "))

square = number ** 2

print("The square of", number, "is", square)
````

## CHALLENGE

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

[PEW] E2.5

````python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit:", fahrenheit)
````
