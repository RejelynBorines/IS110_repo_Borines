#### Exercise 1

In the code below, which will never print regardless of the value for x? Edit the code to correct the error.

```python
if x > 0 :
    print('Above 0')
elif x >  10 :
    print('Above 10')
elif x > 50 :
    print('Above 50')
elif x >  20 :
    print('Above 20')
else :
    print('Something else')
```

The problem is the order of the conditions in the if-elif chain.

Python checks conditions from top to bottom. Once one condition is True, the rest are skipped.If x = 60, the first condition x > 0 is already True, so Python prints "Above 0" and stops checking.

Because of this, the conditions:
x > 10
x > 50
x > 20

Correct is
```python
x = int(input("Enter a number: "))

if x > 50:
    print("Above 50")
elif x > 20:
    print("Above 20")
elif x > 10:
    print("Above 10")
elif x > 0:
    print("Above 0")
else:
    print("Something else")
```

````python
x = int(input("Enter a number: "))

if x > 50:
    print("Above 50")
elif x > 20:
    print("Above 20")
elif x > 10:
    print("Above 10")
elif x > 0:
    print("Above 0")
else:
    print("Something else")
````

#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours <= 40:
    pay = hours * rate
else:
    overtime = hours - 40
    pay = (40 * rate) + (overtime * rate * 1.5)

print("Pay:", pay)
````

## CHALLENGE

WWrite a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		1.75
    [ 82,  85)		2.0
    [ 78,  81)		2.25
    [ 75,  77)		2.5
    [ 0,   74)		5.0


Adapted from [PEW] E3.3
<br>

````python

score = float(input("Enter score (0-100): "))

if score >= 98 and score <= 100:
    grade = 1.0
elif score >= 94 and score < 98:
    grade = 1.25
elif score >= 90 and score < 94:
    grade = 1.5
elif score >= 86 and score < 90:
    grade = 1.75
elif score >= 82 and score < 86:
    grade = 2.0
elif score >= 78 and score < 82:
    grade = 2.25
elif score >= 75 and score < 78:
    grade = 2.5
elif score >= 0 and score < 75:
    grade = 5.0
else:
    grade = "Invalid score"


print("Grade:", grade)
````
