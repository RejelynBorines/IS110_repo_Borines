Create an ndarry object (A) with any values from 1 to 100 in a nested Python List. You can choose your range from the given values (1-100).

* Given that A is an ndarray object, many functionalities in numpy could be accessed by:
    + Make sure to explain each line of codes that you have created.

1. Using a property of the ndarray object
      - e.g. `A.shape`
2. Using a method of the ndarray object
      - e.g. `A.max()`
3. Using a numpy function and passing an ndarray
     - e.g. `np.max(A)`



```python
import numpy as np #it imports the Numpy Libary, in short nickname is np
```

```python
nested_list = [[15, 30, 45], [60, 75, 90]] #nested_list it create a list of list, which represent the how many rows and column
```

```python
A = np.array(nested_list) #converts the orderly nested_list into a NumPy array
```

```python
print("Array Shape (Property):", A.shape) #it shows how many rows and columns are in the array A
```

```python
print("Maximum Value (Method):", A.max()) #it finds the highest number inside the array A
```

```python
print("Average Value (Function):", np.mean(A)) #calculates the average of all numbers in array A
```

```python
print("\nFinal ndarray A:\n", A) #print the final array
```
