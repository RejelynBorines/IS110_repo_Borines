Create an array of values from -5 to + 5. Multiply each value with 2. Print the resulting array.

```python
import numpy as np #it imports the Numpy Libary, in short nickname is np
```

```python
C = np.arange(-5, 6) #create an array of values from -5 to 5, it will starts at -5 and stops before 6 to include +5
```

```python
result = C * 2 #multiply each value by 2 which means it will applies the multiplication to every element at once
```

```python
print("Resulting Array:", result) #display the resulting array
```
