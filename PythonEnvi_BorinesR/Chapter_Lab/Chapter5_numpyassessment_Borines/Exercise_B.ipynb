### Exercise B

Create a 5x5 integer-array filled with ones.
* Check its shape, size, and number of dimensions. Explain briefly the codes you have created by writing a comment beside the code.


```python
import numpy as np #it imports the Numpy Libary, in short nickname is np
```

```python
B = np.ones((5, 5), dtype=int)  #creates 5 rows and 5 columns filled, specifying the data type as 'int'
```

```python
print("Shape:", B.shape) #it tells the form  of the array, it returns 5 rows and 5 columns
```

```python
print("Size:", B.size) #it calculates the total count of all elements stored in the object
```

```python
print("Dimensions:", B.ndim) #this stands for number of dimensions
```

```python
print("\nArray B:\n", B) #will display the created array
```
